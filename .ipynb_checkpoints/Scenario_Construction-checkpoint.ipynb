```python
from metadrive.envs.metadrive_env import MetaDriveEnv
from metadrive.component.vehicle.vehicle_type import DefaultVehicle
from metadrive.component.navigation_module.node_network_navigation import NodeNetworkNavigation
from metadrive.component.navigation_module.trajectory_navigation import TrajectoryNavigation
from metadrive.policy.idm_policy import IDMPolicy, TrajectoryIDMPolicy
from metadrive.policy.expert_policy import ExpertPolicy
from metadrive.utils import generate_gif
from metadrive.scenario.parse_object_state import parse_object_state, get_idm_route, get_max_valid_indicis
from metadrive.component.lane.straight_lane import StraightLane
import copy
import numpy as np



def get_lane_traj(lane):
    len_polygon = len(lane.polygon)
    lane_top = lane.polygon[:len_polygon//2]
    lane_bottom = lane.polygon[len_polygon//2:][::-1]
    traj = []
    for i in range(len(lane_top)):
        traj.append([(lane_top[i][0] + lane_bottom[i][0]) / 2, (lane_top[i][1] + lane_bottom[i][1]) / 2])
    return traj


def sample_traj(start_point, end_point):
    # Define the number of points to interpolate
    num_points = 100
    # Generate a series of t values between 0 and 1
    t_values = np.linspace(0, 1, num_points)
    # Initialize an empty list to hold the interpolated points
    traj = []
    diff_x = end_point[0] - start_point[0]
    diff_y = end_point[1] - start_point[1]
    # Iterate over the t_values
    for i in range(len(t_values)):
        # Compute the x and y coordinates for the interpolated point
        x = start_point[0] + diff_x * t_values[i]
        y = start_point[1] + diff_y * t_values[i]
        # Get the position at the interpolated point and append it to the list
        traj.append(lane.position(x, y))
    return traj

scenario_config = {'map_config': {'type': 'block_sequence', 
                                  'config': 'X', 
                                  # 'lane_width': 10, 
                                  'lane_num': 2, 
                                  # 'start_position': [0, 0],
                                  # 'exit_length': 60,
                                 },
                   'agent_policy':IDMPolicy,
                   'traffic_density':0,
                   'agent_configs':{
                       'default_agent':{
                           'use_special_color':True,
                           'spawn_lane_index':(">>", ">>>", 0),
                           'destination':"1X1_0_"
                       }
                   },
                   'use_render': True,
                  }

env=MetaDriveEnv(scenario_config)

frames = []
curve_traj = []

try:
    env.reset()
    
    cfg=copy.deepcopy(env.config["vehicle_config"])
    cfg["navigation_module"]=NodeNetworkNavigation
    cfg['spawn_lane_index']=("-1X0_1_", "-1X0_0_", 0)
    cfg['destination'] = "1X2_0" # change to different destinations 
    
    all_vehicles = env.engine.get_objects()
    
    for k, v in all_vehicles.items():
        ego_vehicle = v
        break
        
    v = env.engine.spawn_object(DefaultVehicle, 
                                vehicle_config=cfg,
                                # position=[75, 3.5], 
                                # heading=3.14,
                                )
    env.engine.add_policy(v.id, IDMPolicy, v, env.engine.generate_seed())

    for _ in range(200):
        p = env.engine.get_policy(v.name)
        v.before_step(p.act(True))
        p = env.engine.get_policy(ego_vehicle.name)
        env.step(p.act(True))
        frame=env.render(
                         mode="topdown", 
                         window=False,
                         screen_size=(800, 400),
                         camera_position=(60, 10))
        frames.append(frame)
        # update NPC vehicle's state
        v.after_step()
    generate_gif(frames, gif_name="100237.gif")
finally:
    env.close()

from IPython.display import Image
Image(open("100237.gif", "rb").read())
```
